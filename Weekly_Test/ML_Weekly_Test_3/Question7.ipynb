# **7. Plot Any 3 visualizations**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data using numpy
np.random.seed(0)
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))
line_values = np.cumsum(values)

# Bar Chart
plt.figure(figsize=(14, 4))
plt.subplot(1, 3, 1)
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Pie Chart
plt.subplot(1, 3, 2)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Paired(np.arange(len(categories))))
plt.title('Pie Chart')

# Line Chart
plt.subplot(1, 3, 3)
plt.plot(categories, line_values, marker='o', linestyle='-', color='orange')
plt.title('Line Chart')
plt.xlabel('Categories')
plt.ylabel('Cumulative Values')

plt.tight_layout()
plt.show()
```

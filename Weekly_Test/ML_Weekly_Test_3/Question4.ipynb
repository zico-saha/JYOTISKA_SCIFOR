# **4. Implement Support Vector Machine (find SVM metrics till)**

```python
# Importing necessary libraries.
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# Loading the dataset.
DF = pd.read_csv('/content/anemia.csv')
DF.head()
```

```python
# Info and other properties of each columns are already checked in previous questions.
```

```python
# Spliting the data into training and testing sets.
X = DF.iloc[:, 0:5]
y = DF.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardizing or scaling the features.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Training with SVM model.
svm = SVC(kernel='linear', C=1)


svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
```

```python
# Evaluation of the model with different metrices.
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)

accuracy, precision, recall, f1, conf_matrix
```
